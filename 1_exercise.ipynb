# 1章 問題
### 1.1
$$\gamma_{-k,t} = \mbox{Cov}(y_t,y_{t-(-k)}) = E[(y_t - \mu_t)(y_{t-(-k)} - \mu_{t-(-k)})] = E[(y_t - \mu_t)(y_{t+k} - \mu_{t+k})] = E[(y_{t+k} - \mu_{t+k})(y_t - \mu_t)] = \gamma_{k,t+k}$$
よって $\gamma_{-k,t} = \gamma_{k,t+k}$ である。

弱定常過程なので任意の時刻 t で $\gamma_{k,t}$ が等しい。つまり $\gamma_{k,t} = \gamma_{k,t+k}$ である。よって $\gamma_{-k,t} = \gamma_{k,t+k} = \gamma_{k,t}$ が成り立つ。

### 1.2
$$y_t = \mu + \epsilon_t, \epsilon_t \sim W.N.(\sigma^2)$$
この $y_t$ に対し、任意の t と k に対して
$$E(y_t) = \mu$$
$$\mbox{Cov}(y_t,y_{t-k}) = E[(y_t - \mu)(y_{t-k} - \mu)] = \gamma_k$$
が成立することを示せばよい。

ホワイトノイズは期待値が 0 なので、$E(y_t) = E(\mu + \epsilon_t) = E(\mu) + E(\epsilon_t) = \mu$ である。

また $$\mbox{Cov}(y_t,y_{t-k}) = E[(y_t - \mu)(y_{t-k} - \mu)] = E[(\mu + \epsilon_t - \mu)(\mu + \epsilon_{t-k} - \mu)] = E[\epsilon_t \epsilon_{t-k}]$$
$E[\epsilon_t \epsilon_{t-k}]$ はホワイトノイズ $\epsilon$ の自己共分散そのものである。ホワイトノイズは弱定常過程なので自己共分散は一定。
